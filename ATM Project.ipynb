# ATM

```python
class ATMClass:
    def __init__(self, name, pin, balance=0):
        self.name = name
        self.pin = str(pin)
        self.balance = balance

    def verify_pin(self):
        upin = input("Enter your PIN: ")
        if upin == self.pin:
            return True
        else:
            print("❌ Incorrect PIN!")
            return False

    def checkbal(self):
        if self.verify_pin():
            print(f"Hello {self.name}, your current balance is: ₹{self.balance}")

    def withdraw(self):
        amount = float(input("Enter amount to withdraw: ₹"))
        upin = input("Enter your PIN: ")  # ✅ PIN asked after amount
        if upin == self.pin:
            if amount > self.balance:
                print("❌ Insufficient balance!")
            else:
                self.balance -= amount
                print(f"Withdrawal successful! Remaining balance: ₹{self.balance}")
        else:
            print("Incorrect PIN! Transaction cancelled.")

    def deposit(self):
        amount = float(input("Enter amount to deposit: ₹"))
        upin = input("Enter your PIN: ")  # ✅ PIN asked after amount
        if upin == self.pin:
            self.balance += amount
            print(f"Deposit successful! Updated balance: ₹{self.balance}")
        else:
            print("Incorrect PIN! Transaction cancelled.")

    def changepin(self):
        old_pin = input("Enter your old PIN: ")
        if old_pin == self.pin:
            new_pin = input("Enter new PIN: ")
            self.pin = new_pin
            print("PIN changed successfully!")
        else:
            print("Incorrect old PIN!")

def atm():
    user = ATMClass(name="Nikk", pin="1234", balance=10000)

    while True:
        print("\n--------------------------------------------------")
        print("ATM MENU")
        print("--------------------------------------------------")
        print("1. Check Balance")
        print("2. Withdraw Amount")
        print("3. Deposit Amount")
        print("4. Change PIN")
        print("5. Exit")
        print("--------------------------------------------------")

        ch = input("Enter your choice: ")

        if ch == "1":
            user.checkbal()
        elif ch == "2":
            user.withdraw()
        elif ch == "3":
            user.deposit()
        elif ch == "4":
            user.changepin()
        elif ch == "5":
            print("Thank you for using the ATM. Goodbye!")
            break
        else:
            print("Invalid Input. Please try again.")

# Run the program
atm()

```
